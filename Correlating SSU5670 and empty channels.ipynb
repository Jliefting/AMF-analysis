```python
from glob import glob
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib.transforms import Bbox
import matplotlib.font_manager as fm
from scipy.stats import pearsonr
```

```python
files = glob(r'E:\250731_D197198_SSUQ5or670\SSUQ570\*\Zprojected\*.tif')
file_max_values = {}
for file in files:
    img=io.imread(file)
    CY5 = img[...,0].max()
    CY3 = img[...,1].max()
    DAPI = img[...,2].max()
    if any(val == 65535 for val in [CY5, CY3, DAPI]):
        print('saturated channel')
    else:
        file_max_values[file] = {'CY5':CY5, 'CY3':CY3, 'DAPI':DAPI}
```

```python
files = file_max_values.keys()
```

```python
# Assuming `files` is a list of image file paths
fig, axes = plt.subplots(1, len(files), figsize=(5 * len(files), 5))

if len(files) == 1:
    axes = [axes]  # ensure axes is iterable if there's only one file

r2_values_570 = []    

for ax, file in zip(axes, files):
    img = imread(file)
    CY5 = img[..., 0].flatten()
    CY3 = img[..., 1].flatten()
    
    r, _ = pearsonr(CY3, CY5)
    r2 = r**2
    r2_values_570.append(r2)

    ax.scatter(CY3, CY5, s=1, alpha=0.3)
    #ax.set_title(f'{Path(file).parents[1].name}')
    ax.set_xlabel('SSUQ570 Intensity')
    ax.set_ylabel('SSUQ670 Intensity')
    
plt.tight_layout()
plt.show()
```

```python
img = imread(list(files)[0])
CY5 = img[..., 0].flatten()
CY3 = img[..., 1].flatten()

fig, ax = plt.subplots()

ax.scatter(CY3, CY5, s=.1, alpha=0.1, c='sienna')  #

#ax.set_title(f'{Path(file).parents[1].name}')
ax.set_xlabel('SSUQ570 Intensity')
ax.set_ylabel('SSUQ670 Intensity')
plt.plot([0, 50000], [0,50000],color='black')
plt.tight_layout()
#plt.show()

plt.savefig('Channel_correlation_570_empty.png',dpi=600)
```
